```python
#@author- Tejasvi Sharma
import requests
from bs4 import BeautifulSoup
```

```python
#load a page
r = requests.get("https://en.wiktionary.org/wiki/Wiktionary:Main_Page")
```

```python
r.content
```

```python
#declare a soup variable and read in the content
soup = BeautifulSoup(r.content, 'lxml')
```

```python
#printing in a proper format
print(soup.prettify())
```

```python
#find all the anchor (a) tags
t=soup.find_all("div")
#to get all the classes used in div tags
for i in t:
    print(i.get("class"))
    
for i in t:
    print(i.text)
```

```python
#print all the text and the href
for link in soup.find_all("a"):
    print(link.text, link.get("href"))
    print("****************************")
```

```python
#get all the links and display them in proper format
for link in soup.find_all("a"):
    print("<a href='%s'>%s</a>" %(link.get("href"), link.text))
    print("****************************")
```
